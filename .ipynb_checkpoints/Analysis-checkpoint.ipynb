```python
import pickle
with open('ListOfWords.pkl', 'rb') as f:
    ListOfWords = pickle.load(f)

ListOfWords = list(ListOfWords)
```

```python
print(len(ListOfWords))
```

```python
AvgLenght = 0
for i in ListOfWords:
    AvgLenght += len(i)

AvgLenght = AvgLenght / len(ListOfWords)

print(AvgLenght)
```

```python
words = {}

for word in ListOfWords:
    words[word] = {}
```

```python
from fast_autocomplete import AutoComplete
autocomplete = AutoComplete(words=words)
```

```python
import time 
start = time.time()

KeyWords = autocomplete.search(word="doke", max_cost = 3, size = 10)
print(KeyWords)
end = time.time()

print(end - start)
```

```python
def save_object(obj, filename):
    with open(filename, 'wb') as outp:  # Overwrites any existing file.
        pickle.dump(obj, outp, pickle.HIGHEST_PROTOCOL)
# sample usage
save_object(autocomplete, 'company1.pkl')
```

```python
TrendsList = []
```

```python
for i in ListOfWords:
    from pytrends.request import TrendReq

    pytrends = TrendReq(hl='en-US', tz=360) 

    kw_list = [i] # list of keywords to get data 

    pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y')

    data = pytrends.interest_over_time() 
    data = data.reset_index() 

    S = sum(list(data[i]))
    TrendList.append(S)
```

```python
import time

for i in ListOfWords:
    flage = True
    while flage:
        try:
            print("hello")
            from pytrends.request import TrendReq

            pytrends = TrendReq(hl='en-US', tz=360) 

            kw_list = [i] # list of keywords to get data 

            pytrends.build_payload(kw_list, cat=0, timeframe='today 5-y')

            data = pytrends.interest_over_time() 
            data = data.reset_index() 

            S = sum(list(data[i]))
            TrendsList.append(S)
            flage = False
        except Exception as e:
            print("Exception is Handling")
            print(e)
            flage = True
        finally:
            print(flage)
```
